### 3.  You need to provide the Instacart senior stakeholders with descriptive findings about sales. Create a histogram of the “order_hour_of_day” column.

    In a markdown cell beneath your histogram, describe what the histogram demonstrates.


```python
hist= instacart['order_hour_of_day'].plot.hist(bins = 24)
```

##### The histogram illustrates that the majority of orders are concentrated between 9 AM and 6 PM. There is a gradual increase in orders from midnight to around 10 AM, reaching a peak in the number of orders. After 10 AM, the number of orders remains relatively high but stable, with slight decreases until around 5 PM. Following 5 PM, the frequency of orders begins to decline again until midnight.
##### In summary, the histogram indicates that most orders on Instacart are placed during the daytime, with a notable peak around the late morning hours and a gradual decrease towards the evening.

### 7.You’ll also need to explore whether there’s a connection between age and spending power (income). To visualize this relationship, create a scatterplot using the sns.scatterplot() function.

    In a markdown cell beneath your scatterplot, describe what the scatterplot shows.


```python
# Pick the 2 columns we want from the sample

instacartsample3 = small [['age','income']]
```

```python
# Create the scatterplot

scatterplot = sns.scatterplot(data = instacartsample3, x = 'age', y = 'income')
```

##### Generally, there doesn't seem to be a clear correlation between age and income; the points are evenly distributed, showing no discernible pattern of income increasing or decreasing with age. Additionally, starting from age 40, there appears to be a tiered pattern in the income distribution. Up to around 200,000 $, there is a high density of points, suggesting that many customers in this income range have relatively lower earnings. Above 200,000 , the points are more scattered, indicating greater variability in the incomes of customers over 40 years old, with some having significantly higher earnings.
