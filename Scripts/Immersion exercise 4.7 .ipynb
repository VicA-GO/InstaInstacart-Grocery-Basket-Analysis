### If-Statements with User-Defined Functions

```python
# Define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply the function

df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check the values in the new column

df['price_range'].value_counts(dropna = False)
```

```python
# Check the most expensive product within the subset 

df['prices'].max()
```

### If-Statements with For-Loops

```python
#  Create a new column in your ords_prods_merge dataframe that summarizes how busy each day of the week is
```

```python
# To start, you need to know on which day most orders take place

# printing the frequency of the “orders_day_of_week” column

df_ords_prods_merge['orders_day_of_week'].value_counts(dropna = False)
```

```python
#  Create the column “busiest day” with one of three values: “Busiest day,” “Least busy,” and “Regularly busy.” 
# This can be done using a for-loop

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
result
```

```python
# create a new column within your ords_prods_merge dataframe and set it equal to result

df_ords_prods_merge['busiest_day'] = result
```

```python
df_ords_prods_merge['busiest_day'].value_counts(dropna = False)
```

#### 2. Suppose your clients have changed their minds about the labels you created in your “busiest_day” column. Now, they want “Busiest day” to become “Busiest days” (plural). This label should correspond with the two busiest days of the week as opposed to the single busiest day. At the same time, they’d also like to know the two slowest days. Create a new column for this using a suitable method.

```python
# Create the list

result = []

for value in df_ords_prods_merge["orders_day_of_week"]:
  if value == 0 or value == 1:
    result.append("Busiest day")
  elif value == 4 or value == 3:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
# Create the column

df_ords_prods_merge['busiest_days'] = result
```

#### 4. When too many users make Instacart orders at the same time, the app freezes. The senior technical officer at Instacart wants you to identify the busiest hours of the day. Rather than by hour, they want periods of time labeled “Most orders,” “Average orders,” and “Fewest orders.” Create a new column containing these labels called “busiest_period_of_day.”

```python
# Check the hours

df_ords_prods_merge['order_hour_of_day'].value_counts(dropna = False)
```

```python
# Create the list

result = []

for value in df_ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result.append("Most orders")
  elif value in [17, 8, 18, 19, 20, 7, 21, 22]:
    result.append("Average orders")
  else:
    result.append("Fewest orders")
```

```python
# Create the column

df_ords_prods_merge['busiest_period_of_day'] = result
```
